```python
# Import libraries
import spacy
nlp = spacy.load('en_core_web_sm')

# Garden path sentences
gardenpathSentences = ['The raft floated down the river sank.', 'The horse raced past the barn fell.','Mary gave the child a Band-Aid.', 'That Jill is never here hurts.', 'The cotton clothing is made of grows in Mississippi.']
```

```python
# Tokenize each sentence.

doc = [nlp(sentence) for sentence in gardenpathSentences]
token_list = [[token.text for token in doc] for doc in doc]

print(token_list)
```

```python
# Named entity recognition.

gardenpath_words = ' '.join(gardenpathSentences)
nlp_gradenpath_words = nlp(gardenpath_words)
entities = [(ent.text, ent.label_) for ent in nlp_gradenpath_words.ents]
print(entities)
```

```python
# There were only two entities identified by the model.

print(spacy.explain('GPE'))
print(spacy.explain('PERSON'))
```

### PERSON

"Mary" is identified as a person (PERSON) entity. This makes sense, as it is a common personal name. "Jill" is also identified as a person (PERSON) entity. This is appropriate since Jill is a personal name.

### GPE

"Mississippi" is identified as a geopolitical entity (GPE). This also makes sense, as Mississippi is a state in the United States, and geopolitical entities can include countries, states, cities, etc.
